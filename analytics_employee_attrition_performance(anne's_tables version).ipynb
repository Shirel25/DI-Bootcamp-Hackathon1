**Data Preprocessing:**

1. Missing values

```python
# Verify is there are some missing values
employee_att_perform.isnull().sum()
```

There are **no missing values** in the dataset. It is cleaned for analysis.

2. Encoding categorical variables

```python
df_employee_att_perform = employee_att_perform.copy()

df_employee_att_perform.nunique()
# Column that contains a small number of unique values,
# chance are it is categorical
```

```python
le = LabelEncoder()

# Ordinal variables
categorical_columns = ["MaritalStatus", "OverTime", "Gender"]

for col in categorical_columns:
    df_employee_att_perform[col] = le.fit_transform(df_employee_att_perform[col])
    print(f"{col} classes:", le.classes_)

print(df_employee_att_perform[["MaritalStatus", "OverTime", "Gender"]].head())

# One-Hot Encoding nominal variables
df_employee_att_perform = pd.get_dummies(df_employee_att_perform, columns=["JobRole"], drop_first=True)
```

```python
# Verifications
print(df_employee_att_perform[["MaritalStatus", "OverTime", "Gender"]].head())
print("Columns after One-Hot Encoding:", df_employee_att_perform.columns[df_employee_att_perform.columns.str.startswith("JobRole_")].tolist())
```

3. Normalization and standardization

```python
scaler = MinMaxScaler()
df_employee_att_perform[["MonthlyIncome", "Age", "DistanceFromHome"]] = scaler.fit_transform(df_employee_att_perform[["MonthlyIncome", "Age", "DistanceFromHome"]])
```

```python
scaler = StandardScaler()
df_employee_att_perform[["MonthlyIncome", "Age", "DistanceFromHome"]] = scaler.fit_transform(df_employee_att_perform[["MonthlyIncome", "Age", "DistanceFromHome"]])
```

9.** Correlation and Trend Analysis**

```python

le = LabelEncoder() #encoding the attrition to 1 for true and 0 for false.
df_employee_att_perform['Attrition'] = le.fit_transform(df_employee_att_perform['Attrition'])

# Now calculate the correlation matrix
correlation_matrix = df_employee_att_perform[['Attrition', 'Education','EnvironmentSatisfaction', 'JobSatisfaction','PerformanceRating', 'WorkLifeBalance', 'MonthlyIncome']].corr()
print(correlation_matrix)
```

2. Trend Analysis
a. Performance Ratings Across Job Roles and Departments
Method: Calculate the average performance rating for each job role and department, then visualize the trends using bar charts.

```python
# Calculate the average performance rating for each job role and department, then visualize the trends using bar charts.

import matplotlib.pyplot as plt
# Calculate the average performance rating for each job role and department
average_performance = employee_att_perform.groupby(['JobRole', 'Department'])['PerformanceRating'].mean().reset_index()

# Visualize the trends using bar charts
plt.figure(figsize=(12, 6))
sns.barplot(x='JobRole', y='PerformanceRating', hue='Department', data=average_performance)
plt.title('Average Performance Rating by Job Role and Department')
plt.xlabel('Job Role')
plt.ylabel('Average Performance Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

3. Cross-Tabulations
a. DistanceFromHome by JobRole and Attrition
Method: Create a cross-tabulation table to show the count of employees based on DistanceFromHome categorized by JobRole and Attrition status.
b. MonthlyIncome by Education and Attrition
Method: Generate a cross-tabulation to show average MonthlyIncome categorized by Education level and Attrition status.

```python
# Create a cross-tabulation table to show the count of employees based on DistanceFromHome categorized by JobRole and Attrition status.

crosstab = pd.crosstab(index=[employee_att_perform['JobRole'], employee_att_perform['Attrition']],
                       columns=employee_att_perform['DistanceFromHome'])

crosstab
```

```python
# Generate a cross-tabulation to show average MonthlyIncome categorized by Education level and Attrition status.

crosstab_income = pd.pivot_table(data=employee_att_perform, values='MonthlyIncome', index='Education', columns='Attrition', aggfunc='mean')
crosstab_income
```

```python
# Heatmaps: Use heatmaps to visualize the relationships in the cross-tabulations, showing the intensity of counts or averages.
# Heatmaps for cross-tabulations
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Distance From Home by Job Role and Attrition')
plt.xlabel('Distance From Home')
plt.ylabel('Job Role and Attrition')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(8, 6))
sns.heatmap(crosstab_income, annot=True, fmt=".2f", cmap="YlOrRd")
plt.title('Average Monthly Income by Education and Attrition')
plt.xlabel('Attrition')
plt.ylabel('Education')
plt.show()
```
